```python
data_h = df.copy()
data_h = data_h.set_index('Country')      # Setting countries as index to create a new data frame with the same index

countries_region = pd.DataFrame(index=data_h.index)      # Creating Data Frame with index countries 


# Loops for 0's and 1's to identify to which region belongs each country.


W_Europe = []                   # Accumulator Pattern
for r in df['Region']:
    if r == 'Western Europe':   # if the region is equal to Western Europe, add 1 in the List W_Europe
        W_Europe.append(1)
    else:                       # else add 0
        W_Europe.append(0)

        
E_Europe = []   
for r in df['Region']:
    if r == 'Eastern Europe':
        E_Europe.append(1)
    else:
        E_Europe.append(0)

        
L_America = []
for r in df['Region']:
    if r == 'Latin America':
        L_America.append(1)
    else:
        L_America.append(0)

        
N_America = []
for r in df['Region']:
    if r == 'North America':
        N_America.append(1)
    else:
        N_America.append(0)  

Africa = []
for r in df['Region']:
    if r == 'Africa':
        Africa.append(1)
    else:
        Africa.append(0)   

        
Asia = []
for r in df['Region']:
    if r == 'Asia-Pacific':
        Asia.append(1)
    else:
        Asia.append(0)
        
# Filling DataFrame

countries_region['Western_Europe'] = W_Europe
countries_region['Eastern_Europe'] = E_Europe
countries_region['Latin_America'] = L_America
countries_region['North_America'] = N_America
countries_region['Africa'] = Africa
countries_region['Asia'] = Asia
```
